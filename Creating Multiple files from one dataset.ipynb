## Importing libraries

```python
import os
import pandas as pd
```

## Reading files

```python
A=pd.read_excel('sales_data.xlsx',index_col="Row ID")
A.head()
```

```python
A.info()
```

```python
A['M_Y']=A['Order Date'].dt.strftime("%m-%Y")
```

```python
A.head()
```

```python
gb=A.groupby('M_Y') # Grouping according to the year and month
```

```python
## Creating Empty list and adding values according to the groups
ls=[]
for each in gb.groups:
    ls.append(gb.get_group(each))
```

```python
#Creating Excel files according to the month and year combination
for each in ls:
    filename=each['M_Y'].iloc[0]
    each.to_excel("{}.xlsx".format(filename),index=False)
```
